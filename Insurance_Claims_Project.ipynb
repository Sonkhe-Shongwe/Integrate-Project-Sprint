# Introduction

**Project Overview:**
- In South Africa, despite a relatively strong economic position within the continent, the insurance industry faces significant challenges. These include low penetration rates, affordability concerns, a lack of awareness about the importance of insurance, and a general skepticism towards insurance products among the population. Our project aims to tackle these challenges head-on by introducing innovative solutions that make insurance more accessible, affordable, and relevant to the South African population.

## Defining a Problem Statement

**Background:**

- Africa, with South Africa being a leading economy, represents a vast untapped market for insurance services. However, several barriers prevent the majority from accessing these services, notably in the realms of affordability, awareness, and trust. Furthermore, the specific needs and risks of the South African population are not always adequately addressed by existing insurance products.

**Problem Statement:**

- In South Africa, a significant portion of the population remains uninsured or underinsured due to a combination of high insurance premiums, low financial literacy, and a lack of trust in insurance providers. This situation exposes individuals and businesses to considerable risk in the event of accidents, health issues, or natural disasters. Moreover, the existing insurance models and products often fail to meet the unique needs and preferences of diverse segments of the South African market, further exacerbating the gap in insurance coverage.

**Project Objectives:**
1. To conduct an in-depth analysis of the barriers to insurance access in South Africa, focusing on affordability, awareness, and trust.
2. To develop and pilot innovative insurance products that are affordable, understandable, and relevant to the needs of the South African market.
3. To utilize technology and data analytics to create personalized insurance offerings and streamline the claims process, improving trust and customer satisfaction.
4. To launch an educational campaign aimed at increasing financial literacy and awareness of the benefits of insurance, specifically targeting underserved communities.

## 5.6 High Risk Hobbies FLag

```python
# Define high-risk hobbies
high_risk_hobbies = ['bungie-jumping', 'base-jumping', 'skydiving', 'paintball']

# Define medium-risk hobbies
medium_risk_hobbies = ['kayaking', 'basketball', 'polo', 'cross-fit', 'exercise', 'dancing']

# Define low-risk hobbies
low_risk_hobbies = ['sleeping', 'reading', 'board-games', 'golf', 'camping', 'movies', 'hiking', 'yachting', 'chess', 'video-games']
```

```python
# Define a function to assign risk levels to hobbies
def assign_hobby_risk_level(hobby):
    """
    Assigns a risk level to a given hobby.

    Args:
    - hobby (str): The hobby to assign a risk level to.

    Returns:
    - str: The risk level assigned to the hobby.
    """
    if hobby in high_risk_hobbies:
        return 'High Risk'
    elif hobby in medium_risk_hobbies:
        return 'Medium Risk'
    else:
        return 'Low Risk'

# Apply the function to create a new column 'hobby_risk_level'
df_insurance['hobby_risk_level'] = df_insurance['insured_hobbies'].apply(assign_hobby_risk_level)
```

```python
# Check Hobbies and Hobby risk level
df_insurance[['insured_hobbies','hobby_risk_level']].tail()
```

```python
# Group the data by 'hobby_risk_level' and 'incident_severity', then count the occurrences
severity_hobby_correlation = df_insurance.groupby(['hobby_risk_level', 'incident_severity']).size().unstack(fill_value=0)

# Plot the bar chart to show incident severity by hobby risk level
severity_hobby_correlation.plot(kind='bar')

# Set title and axis labels
plt.title('Incident Severity by Hobby Risk Level', fontweight='bold')
plt.xlabel('Hobby Risk Level', fontweight='bold')
plt.ylabel('Incident Count', fontweight='bold')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title='Incident Severity')  # Add legend with title
plt.show()
```
